(b) Aggregate and query

<font size = "3">

- Import the "data_raw/sprint_results.csv" file
- Subset the observations for 'constructorId == 3' using ```.query()```
- Obtain the maximum of "laps" by "driverId" <br> 
 using ``` .groupby().agg() ```
- Subset the drivers with a maximum of laps $\ge 10$
- Store the results of all these operations in a dataframe <br>
```sprint_results_agg ```

````python
# Write your own code

sprint = pd.read_csv("data_raw/sprint_results.csv")

sprint_results_laps = (sprint.query("constructorId == 3")
                      .groupby("driverId")
                      .agg(max_laps = ("laps", "max")))

sprint_results_agg = sprint_results_laps.query("max_laps >= 10")
display(sprint_results_agg)
````

(c) Recode a numeric colum

<font size = "3">

- Import the "data_raw/races.csv" file
- Recode the "year" column into a new column "year_brackets" <br>
with the following categories

$\qquad$ ``` ["1950-1965","1966-1980","1980-onwards"] ```

$\qquad$ HINT: Use the "pd.cut()" command.

````python
# Write your own code

races = pd.read_csv("data_raw/races.csv")

bins1 = [1950, 1965, 1980, 2024]
labels1 = ["1950-1965","1966-1980","1981-onwards"]

races["year_brackets"] = pd.cut(races["year"],
                                bins=bins1,
                                right=True,
                                labels = labels1)

# correction:
# bins1 = [1949, 1965, 1980, float("inf")]
# bc the left bracket is not inclusive in pd.cut
````

(d) Aggregate and sort

<font size = "3">

- Import the "data_raw/sprint_results.csv" file
- Obtain the maximum of "laps" by "driverId" <br> 
using ``` .groupby().agg() ```
- Sort the teams in descending order using of <br>
the maximum of laps using ``` .sort_values() ```

````python
# Write your own code

sprint2 = pd.read_csv("data_raw/sprint_results.csv")

laps_agg = (sprint2.groupby("driverId")
            .agg(maximum_laps = ("laps", "max"))
            .sort_values(by="maximum_laps", ascending=False))

laps_agg
````
